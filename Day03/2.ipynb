```python
#example data frame plot
import numpy as np
import pandas as pd
df1=pd.DataFrame(np.random.rand(10,4),columns=('col1','col2','col3','col4'))
print(df1)
```

```python
df1.plot()
```

```python
df1.plot(kind='hist')
```

```python
df1.plot(subplots=True,figsize=(8,8))
```

```python
df1.plot(kind='bar')
```

```python
df1.plot(kind='line')
```

```python
df1.plot(kind='pie',y='col3')
```

```python
df1.plot(kind='box')
```

```python
df1.plot(kind='kde')
```

```python
df1.plot(kind='area')
```

```python
df1.plot(kind='scatter', x=2 ,y='col2')
```

```python
df1.plot(kind='hexbin',x='col2',y='col1',gridsize=6)
```
