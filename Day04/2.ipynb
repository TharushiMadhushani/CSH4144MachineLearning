```python
#Scatter plot and mat plot

import pandas as pd
#create data frame
df=pd.DataFrame({'x':[5,10,15,20,25,30,35],
                'y':[5,10,15,20,25,30,35]})

print(df)
```

```python
#crate a scatter plot

df.plot.scatter(x='x',y='y')
```

```python
#customize the scatter plot
df.plot.scatter(x='x',y='y',s=20,c='green')
```
